```python
import pandas as pd
df = pd.read_csv("result.csv", index_col=[0])
df['price'] = df['price'].str.replace(' ', '').astype(int)
df[['smartphone_brand','smartphone_model']] = df["title"].str.split(" ", 1, expand=True)
df = df.drop_duplicates(subset=['title', 'price']).reset_index(drop=True)
operations = {col_name:('first' if col_name != 'price' else 'mean') for col_name in df}
df = df.groupby('title', as_index=False).agg(operations)
df["smartphone_id"] = df.index + 1
df['price'] = df['price'].astype(int)
df = df[['smartphone_id', 'smartphone_brand', 'smartphone_model', 'price']]
df
```
